# 과적합 피하기 - 초음파광물 사례
## 학습셋과 테스트셋으로 구분

```python
import pandas as pd

df = pd.read_csv('dataset/sonar.csv', header=None)
df.head()
```

```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
dataset = df.values
X = np.array(dataset[:,0:60], dtype=np.float64)
Y_obj = dataset[:,60]
```

```python
# 문자열 변환
e = LabelEncoder().fit(Y_obj)
Y = np.array(e.transform(Y_obj), dtype=np.float64)
```

```python
# 학습셋과 테스트셋의 구분
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, 
                                                    random_state=seed)
```

```python
# 모델 설정
model = Sequential([
    Dense(24, input_dim=60, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
# 모델 컴파일 
model.compile(loss='binary_crossentropy',   # mean_squared_error
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=130, batch_size=5)
```

```python
# 테스트셋에 모델 적용 
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test, verbose=2) [1]))
```
